```python
import pandas as pd
import plotly as py
import plotly.graph_objects as go
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
coins = cg.get_coins_list()
for coin in coins:
    if coin['name'] == 'Bitcoin':
        print(coin)
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'usd', days = 30)
```

```python
BTCdata = pd.DataFrame(bitcoin_data)
BTCdata.head()
type(BTCdata['prices'].iloc[0])
```

```python
BTCprice = pd.DataFrame(
    BTCdata['prices'].tolist(),
    columns=['TimeStamp','Price']
)
```

```python
BTCprice.head()
```

```python
BTCprice['Date'] = pd.to_datetime(BTCprice['TimeStamp'], unit='ms')
BTCprice.head()
```

```python
BTCcandle = BTCprice.groupby(BTCprice.Date.dt.date).agg({'Price': ['min', 'max', 'first', 'last']})
```

```python
BTCcandle.tail()
```

```python
fig = go.Figure(
    data=[go.Candlestick(x = BTCcandle.index,
                                open = BTCcandle['Price']['first'],
                                high = BTCcandle['Price']['max'],
                                low = BTCcandle['Price']['min'],
                                close = BTCcandle['Price']['last']
                        )
          ]
)
```

```python
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title = 'Date', yaxis_title = 'Price (USD)', title = 'Bitcoin 30 day candlestick')
fig.show()
```

```python
py.offline.plot(fig, filename='./bitcoin_chart.html', auto_open=True)
```
